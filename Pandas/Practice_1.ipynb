Indexing

```python
s=pd.Series([12,30,50,70,90])
print(s)
print(s[0])
print(s[4])

print(s[0:2])  #start(indluded):stop (excluded) : step value(values to jump)

print(s[0:4:2])
```

```python
# iloc--> location based indexing (it gives value in that particular location)
s.iloc[3]
```

```python
# want multiple values
s.iloc[[1,2,4]]
```
