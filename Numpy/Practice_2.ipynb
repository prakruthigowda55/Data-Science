Array Reshaping--> Reshape,Ravel,Flatten

```python
import numpy as np
```

```python
#reshape
arr=np.array([1,2,3,4,5,6])
reshaped=arr.reshape(2,3)
print(reshaped)
```

```python
reshape2=reshaped.reshape(3,2)
print(reshape2)
```

```python
# reval-->convert 1D array
ravel=reshaped.ravel()
print(ravel)
```

```python
ravel[0]=100
print(ravel) # it is a view
print(reshaped)
```

```python
#  flatten --> convert to 1D array
# returns a copy of the array

flat=reshaped.flatten()
print(flat)

```

```python
flat[1]=10
print(flat)
print(reshaped)
```
