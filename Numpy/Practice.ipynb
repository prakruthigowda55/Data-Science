```python
# create arrays
range = np.arange(1,10) #last number not included
range
```

```python
range = np.arange(1,10,4) # start stop step
range
```

```python
# linspace
arr=np.linspace(1,2,5) #start ,stop, number of values
arr
```

```python
# logarithmic scale array
arr=np.logspace(1,3,3) # 10^1--10^3 need 3 points
arr
```
