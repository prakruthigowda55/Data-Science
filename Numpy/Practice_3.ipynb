```python
import numpy as np
```

Repeating

```python
# repeats each elemnt
arr=np.array([1,2,3])
print(np.repeat(arr,3))
```

Tile

```python
# repeate the whole array
print(np.tile(arr,2))
```

Aggregrate Functions --> summerize the entire array

```python
arr=np.array([1,2,3])

print(np.sum(arr)) # sum of arrays
print(np.mean(arr)) # average of array
print(np.min(arr)) # minimum val
print(np.max(arr)) # max val
print(np.median(arr)) # middle val
print(np.std(arr)) # standard diviation
print(np.var(arr)) #variance --> sq od std
```

```python
matrix=np.array([[1,2,3],[4,5,6]])

print(np.sum(matrix,axis=1))
```

Cumulative operations--> running total

```python
arr=np.array([1,2,3])

print(np.cumsum(arr))
print(np.cumprod(arr)) #product
```

Conditional Based choices

```python
# Where

arr=np.array([1,2,3])
print(arr)

result = np.where(arr<=2, "low" , "high")
print(result)
```

```python
# argwhere: row-column positions for 2d array
arr=np.array([10,20,40,50,60,12])
print(np.argwhere(arr>30))
```

```python
# logical and

mask=np.logical_and(arr>20 , arr<60)
print(mask)
print(arr[mask])
```

```python
# logical or
mask=np.logical_or(arr<20, arr>70)
print(mask)
print(arr[mask])
```

```python
# logical not
mask=np.logical_not(arr>50)
print(mask)
```

Broadcasting--> strecting the shape of smaller arrays so that they match the bigger arrays

```python
a=np.array([1,2,3,4,5])
b=np.array(5) # due to broadcasing [5,5,5,5,5]
print(a+b)
```

```python
image=np.array([[200,250],[300,400]])
print(image)

brightness=image+50
print(brightness)
```

```python
a=np.array([[1,2,3]])
b=np.array([[1],[2],[3]])
print(b.shape)
print(a.shape)
ans=a+b
print(ans)
print(ans.shape)
```

```python
a=np.array([[1,2,3],[1,4,5]])
b=np.array([[1],[2],[3]])
print(b.shape)
print(a.shape)
ans=a+b
print(ans)
print(ans.shape)
```

Vectorization

```python
# np.vectorize()--> convert a regular function to be applied on a array
def square(x):
    return x*x
vfunc=np.vectorize(square)

arr=np.array([2,3,4,5])
print(vfunc(arr))
```

missing values

```python
# np.nan-->not a number(used to represnt the missing or undefined values )
a=np.array([1,2,4,np.nan,5])
print(a)
```

```python
# np.inf and -np.inf--> represnt the + and - infinit value
a=np.array([1,2,4,np.inf,5])
print(a)
```

```python
# np.isnan
# np.isinf
# np.isfinite
# these are the functions used to detect any null values infinte values or finite values
```

```python
arr=np.array([1,2,np.nan,-33,np.inf])
print(np.isnan(arr))
print(np.isinf(arr))
print(np.isfinite(arr))
```

```python
# np.nan_to_num() replace non to number
new=np.nan_to_num(arr)
print(new)
```
